```python
url_check = 'http://www.lagou.com/jobs/positionAjax.json?gj=1-3%E5%B9%B4&px=default&city=%E5%85%A8%E5%9B%BD&first=false&pn=1&kd=%E6%95%B0%E6%8D%AE%E5%88%86%E6%9E%90%E5%B8%88'
```

```python
data = urllib.urlopen(url_check).read()
data1= json.loads(data)
```

```python
np_total = data1['content']['totalPageCount']
```

```python
i = 1
urls_stack = list()
while i<=np_total:
    pn = i
    url = 'http://www.lagou.com/jobs/positionAjax.json?gj=1-3%E5%B9%B4&px=default&city=%E5%85%A8%E5%9B%BD&first=false&pn='+str(pn)+'&kd=%E6%95%B0%E6%8D%AE%E5%88%86%E6%9E%90%E5%B8%88'
    urls_stack.append(url)
    i += 1
#urls_stack   
```

```python
##2.Read the urls and pickup all positionID
```

```python
##(2) Pack up all position IDs
def pos_stack(data1):
    for i in range(0,len(data1["content"]['result'])):
        pos_id_stack.append(data1["content"]['result'][i]['positionId'])
```

```python
##(2)Literate all urls informations
def urls_reading(urls_stack):
    pos_id_stack = list()
    for link in urls_stack:
        data = urllib.urlopen(link).read()
        data1= json.loads(data)
        pos_stack(data1)
```

```python
#(1)Put all position ids into a stack
def url_process(urls_stack):
    urls_reading(urls_stack)
    ids = np.unique(pos_id_stack)
    clean_ids = [ x for x in ids if isinstance(x,int) ]
    return clean_ids
```

```python
def ids_urls(position_ids):
    position_id_urls = list()
    for ids in position_ids:
        position_id_urls.append('http://www.lagou.com/jobs/'+str(ids)+'.html')
    return position_id_urls
```

```python
position_ids = url_process(urls_stack)
```

```python
position_id_urls = ids_urls(position_ids)
```
