```python
#coding=utf-8
import urllib2
from bs4 import BeautifulSoup as bs
from __future__ import print_function 
import pandas as pd
import numpy as np
import pickle
```

```python
def infor_load(url):
    content = urllib2.urlopen(url).read()
    text = bs(content)
    return text

def print_code(code):
    for letter in code:
        print (letter,end='')
        
def get_info(text):   
    #name_company
    name_company = text.title.contents
    #Industry
    uni_str=text.findAll(name='li')[5].contents[2]
    uni_str
    #Numbers of people
    num_people = text.findAll(name='li')[6].contents[2]
    #Salary
    salary = text.findAll(name='span')[3].contents[0]
    #Job Description
    jd = text.findAll("dd")[1].contents[3].contents
    #Location
    location = text.findAll("dd")[5].contents[9].contents
    #Financing
    Financing = text.findAll('li')[8].contents[1]
    #Job Description
    jd = text.findAll("dd")[1].contents[3].contents
    ##Put them in Data Frame 
    info = [name_company,uni_str,num_people,salary,location,Financing,jd]
    
    return info  

def get_jd(text):
    jd = text.findAll("dd")[1].contents[3].contents
    return jd
```

```python
url = 'http://www.lagou.com/jobs/1447355.html'
url2 = 'http://www.lagou.com/jobs/511234.html'
url3 = 'http://www.lagou.com/jobs/444288.html'
```

```python
urls = [url,url2,url3]
```

```python
##Read the data of all lists of urls
f = open('urls_list.txt',"r")
abc = f.read()
f.close()
for line in abc.split('\n'):
    print(line)
```

```python
def load_urls(urls):
    if 'r_hold' in vars():
        del r_hold
    text = infor_load(urls[0])
    df = get_info(text)
    r = pd.DataFrame(pd.Series(data=df)).T
    jd = list()
    jd.append(get_jd(text))
    for url in urls[1:]:
        try:
            text = infor_load(url)
            df = get_info(text)
            r_hold = pd.DataFrame(pd.Series(data=df)).T
            r = pd.concat([r,r_hold])
            jd.append(get_jd(text))
        except Exception:
            print(Exception)
            continue
    r.columns = ["Company","Industry","People","Salary","Location","Financing","JD"]
    return {'job_basic':r, 'job_jd':jd}
```
